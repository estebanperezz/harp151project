```python
import tkinter as tk
import re
from tkinter import*
import requests
from tkinter import ttk
from tkinter.messagebox import showinfo
from PIL import Image, ImageTk
from tkinter import messagebox
import pandas as pd 
import csv
import os
## Starts a window 
root = tk.Tk() 
root.title('Project Window Demo')
## Gives the window the size
root.geometry('700x500+50+50')
## displays the hello world message
message =tk.Label(root,text="Hello World",bg="green",fg="white")
message.pack(fill=tk.X) 
## Puts Netflix logo into box
##load = Image.open("netflix2.jpeg")
##render = ImageTk.PhotoImage(load)
##Puts hulu logo into box
##load = Image.open("hulu.jpeg")
##render2 = ImageTk.PhotoImage(load)
##box1 = tk.Label(root,image =render)

##box1.pack(ipadx=20, ipady=20, expand=True, side=tk.LEFT)

# box 2
##box2 = tk.Label(root,image=render2)
##box2.pack(ipadx=20, ipady=20, expand=True, side=tk.RIGHT)
selected= tk.StringVar()
Wtype=tk.StringVar()
radiotv=ttk.Radiobutton(root,text="TV",value=2, variable=Wtype)
radioMovie=ttk.Radiobutton(root,text="Movie",value=1, variable=Wtype)
radiotv.pack(side=tk.LEFT)
radioMovie.pack(side=tk.LEFT)
r1 = ttk.Radiobutton(root, text='Action', value=28, variable=selected)
r2 = ttk.Radiobutton(root, text='Sifi', value=878, variable=selected)
r3 = ttk.Radiobutton(root, text='Drama', value=18, variable=selected)
listbox=Listbox(root,height=15,width=30)
listbox.insert(END,"Select TV or Movie and Genre")
myRating=[]
mySummary=[]
def selectedsum():
    for item in listbox.curselection():
        messagebox.showinfo("Summary" , mySummary[item])
        
def selectedcurs():
    for item in listbox.curselection():
        messagebox.showinfo("Rating",str(myRating[item]))       
def selectedget():
    if selected.get()=="" or Wtype.get()=="":
        messagebox.showerror("Error", "Choose Movie or Tv and Select Genre")
        button3.pack_forget()
        button4.pack_forget()
        listbox.delete(0,END)
    else:
        pattern="[a-zA-Z0-9]"
        if len(myRating)>1:
            myRating.clear()
            mySummary.clear()
        listbox.delete(0,END)
        i = 1
        if Wtype.get()=="1":
            while i<20:
                json_file=requests.get(f"http://api.themoviedb.org/3/discover/movie?api_key=2b0268899e7722c303709443c41c21f4&with_original_language=en&language=en-US&region=US&sort_by=popularity.desc&include_adult=false&include_video=false&with_genres={selected.get()}&page={i}").json()
                i+=1
                for elem in json_file['results']:
                    title = elem.get('original_title')
                    if(re.search(pattern,title)):
                        voteavg = elem.get('vote_average')
                        myRating.append(voteavg)
                        votenum = elem.get('vote_count')
                        summary =elem.get('overview')
                        mySummary.append(summary)
                        listbox.insert(END,title)
                    
    ##Since the movie genre id and tv genre ids are differnt i reassigned the 
    ##values for the search to work
        elif Wtype.get()=="2":
            genreVar=0 
            if selected.get()=="28":
                genreVar=10759
                print(genreVar)
            elif selected.get()=="878":
                genreVar=10765
            elif selected.get()=="18":
                genreVar=18
        
            while i<20:
                json_file=requests.get(f"http://api.themoviedb.org/3/discover/tv?api_key=2b0268899e7722c303709443c41c21f4&with_original_language=en&language=en-US&region=US&sort_by=popularity.desc&include_adult=false&include_video=false&with_genres={genreVar}&with_watch_providers=8&watch_region=US&page={i}").json()
                i+=1
                for elem in json_file['results']:
                    title = elem.get('original_name')
                    if(re.search(pattern,title)):
                        idnum=elem.get('id')
                        voteavg = elem.get('vote_average')
                        myRating.append(voteavg)
                        votenum = elem.get('vote_count')
                        summary =elem.get('overview')
                        mySummary.append(summary)
                        listbox.insert(END,title)
        button3.pack() 
        button4.pack()
        
button2=ttk.Button(root,text="Get Option",command=selectedget)
button3=ttk.Button(root,text="Get Rating",command=selectedcurs)
button4=ttk.Button(root,text="Get Summary",command=selectedsum)
r1.pack()
r2.pack()
r3.pack()   
listbox.pack()
button2.pack()
root.mainloop()
```
