```python
##Program  will get list of available streaming services
import requests
import csv 
import pandas as pd
from matplotlib import pyplot as plt 
import numpy as np 
import urllib.request
import json

api_Key = "l4T5GRmxtJ1uElzIWngdXQaGS4o5fgDn3nrWMqkl"

with urllib.request.urlopen(f"https://api.watchmode.com/v1/sources/?apiKey={api_Key}") as url:
    data = json.loads(url.read().decode())
    data_file=open ("streamingservices.csv", "w", newline="", encoding="utf-8")
    csv_writer = csv.writer(data_file)
    csv_writer.writerow(["id#", "Name", "Type"])
    for element in data:
        idnum = element["id"]
        name = element["name"]
        types = element["type"]
        csv_writer.writerow([idnum,name,types])
data_file.close()
file = pd.read_csv("streamingservices.csv")
file
```
